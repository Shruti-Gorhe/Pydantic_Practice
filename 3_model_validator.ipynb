```python
from pydantic import BaseModel, EmailStr, model_validator
from typing import Dict, List

class Patient(BaseModel):
    name: str
    email: EmailStr
    age: int
    weight: float
    married: bool
    allergies: List[str]
    contact_details: Dict[str, str]

    @model_validator(mode='after')
    def validate_patient(self):

        # Business Rule:
        # If age > 60, emergency_contact must be present

        if self.age > 60:
            if 'emergency_contact' not in self.contact_details:
                raise ValueError(
                    "Patients older than 60 must provide an emergency contact number."
                )

        return self


def update_patient_data(patient: Patient):
    print(patient.name)
    print(patient.age)
    print(patient.weight)
    print(patient.married)
    print(patient.allergies)
    print("updated in database!!")


patient_info = {
    "name": "Shruti",
    "email": "shruti@hdfc.com",
    "age": 65,
    "weight": 55.5,
    "married": False,
    "allergies": ["peanuts"],
    "contact_details": {
        "phone": "123-456-7890",
        "emergency_contact": "987-654-3210"
    }
}

patient1 = Patient(**patient_info)

update_patient_data(patient1)
```
