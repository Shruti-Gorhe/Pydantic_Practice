```python
#Computed Fields: Fields that are derived from other fields in the model. They are not explicitly provided by the user but are computed based on the values of other fields.

from pydantic import BaseModel, EmailStr, computed_field
from typing import Dict, List

class Patient(BaseModel):
    name: str
    email: EmailStr
    age: int
    weight: float      # kg
    height: float      # meters
    married: bool
    allergies: List[str]
    contact_details: Dict[str, str]

    @computed_field
    @property
    def bmi(self) -> float:
        return round(self.weight / (self.height ** 2), 2)


def update_patient_data(patient: Patient):
    print(patient.name)
    print(patient.age)
    print(patient.weight)
    print(patient.height)
    print(patient.bmi)
    print(patient.married)
    print(patient.allergies)
    print("updated in database!!")


patient_info = {
    "name": "Shruti",
    "email": "shruti@hdfc.com",
    "age": 22,
    "weight": 55.5,
    "height": 1.65,
    "married": False,
    "allergies": ["peanuts"],
    "contact_details": {
        "phone": "123-456-7890"
    }
}

patient1 = Patient(**patient_info)

update_patient_data(patient1)

print(patient1.model_dump())
```
