```python
#how does pydantic work
from os import name

from pydantic import BaseModel,EmailStr,AnyUrl,Field 
from typing import Annotated, List
from typing import Dict ,Optional
from pydantic import BaseModel, EmailStr, AnyUrl, Field
from typing import Annotated, List, Dict, Optional

class Patient(BaseModel):
    #how does pydantic work

    class Patient(BaseModel):
        name: str = Field(
            min_length=1,
            max_length=50,
            title="Name of the patient",
            description="The name of the patient",
            example="John Doe",
        )
        email: EmailStr
        age: int
        weight: float = Field(gt=0)
        married: Optional[bool]
        allergies: Optional[List[str]] = None
        contact_details: Dict[str, str]

    def insert_patient_data(patient: Patient):
        print(patient.name)
        print(patient.email)
        print(patient.age)
        print(patient.weight)
        print(patient.married)
        print(patient.allergies)
        print(patient.contact_details)
        print("inserted into database!!")

    def update_patient_data(patient: Patient):
        print(patient.name)
        print(patient.email)
        print(patient.age)
        print(patient.weight)
        print(patient.married)
        print(patient.allergies)
        print("updated in database!!")

    patient_info = {
        "name": "Shruti",
        "email": "shruti@example.com",
        "age": 22,
        "weight": 55.5,
        "married": False,
        "allergies": ["peanuts"],
        "contact_details": {"phone": "123-456-7890"},
    }
    patient1 = Patient(**patient_info)

    insert_patient_data(patient1)
    update_patient_data(patient1)
    age:int
    weight:float = Field(gt=0)
    married:Optional[bool]
    allergies:Optional[List[str]] =None  #if we might have written only list then it would have accepted any type of data in the list but here we have specified that the list should contain only strings
    contact_details:Dict[str,str] #if we might have written only dict then it would have accepted any type of data in the dictionary but here we have specified that the dictionary should contain only strings as keys and values

def insert_patient_data(patient: Patient):# patient object is passed as an argument to the function
    print(patient.name)
    print(patient.age)
    print(patient.weight)
    print(patient.married)
    print(patient.allergies)
    print(patient.contact_details)
    print("inserted into database!!")

def update_patient_data(patient: Patient):# patient object is passed as an argument to the function
    print(patient.name)
    print(patient.age)
    print(patient.weight)
    print(patient.married)
    print(patient.allergies)
    print("updated in database!!")

patient_info = {"name":"Shruti" , "email":"shruti@example.com", "age":22 , "weight":55.5 , "married":False , "allergies":["peanuts"] , "contact_details":{"phone":"123-456-7890"}} #dictionary with patient information
patient1 = Patient(**patient_info)  #unpacking the dictionary into the Patient class

insert_patient_data(patient1) #passing the patient object to the function
update_patient_data(patient1) #passing the patient object to the function

```
